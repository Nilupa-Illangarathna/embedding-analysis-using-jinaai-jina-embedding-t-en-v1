```python
import time
from IPython.display import clear_output, display

def print_two_lines(step):
    print(f"step: {step}")
    time.sleep(0)
    clear_output(wait=True)

def Print_The_State(step):
    # Clear the output cell in Colab
    clear_output(wait=True)

    # Print the number of characters on the first line
    print_two_lines(step)
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import time
from IPython.display import clear_output, display

# Load the SentenceTransformer model
model = SentenceTransformer('jinaai/jina-embedding-t-en-v1')
```

## **Start from here**

```python
# Initialize a global counter
count = 0
```

```python
# df = df_selected_columns.iloc[0:400000]
df = data.iloc[400000:515737]
```

```python
print(len(df))
display(df.head(2))
```

```python
def get_sentence_embeddings(passage):
    global count
    count += 1
    result_monitor = f"""Progress: {count}
Completed row count: {count}"""
    print_two_lines(result_monitor)

    # List of sentences
    sentences = [sentence.strip() for sentence in passage.split('.') if sentence.strip()]

    # Encode the sentences
    embeddings = model.encode(sentences, convert_to_tensor=True)

    return [{'sentence': sentence, 'embedding': embedding.tolist()} for sentence, embedding in zip(sentences, embeddings)]

# Apply the function to each row of the DataFrame
df['Positive_Review_Embeddings'] = df['Positive_Review'].apply(lambda x: get_sentence_embeddings(x))

# Print the overall progress count
print(f"\nTotal rows processed: {count}")

# Display the DataFrame with the new columns
display(df.head(3))
```
